```python
import numpy as np
import matplotlib.pylab as plt
from dataset import Dataset
from soft_dtw import SoftDTW
from scipy.optimize import minimize
from soft_dtw import SoftDTW
from softdtw_barycenter import softdtw_barycenter
```

```python
ds = Dataset("ECG200")
X_train, y_train, X_test, y_test = ds.load_dataset()
n_series = 10
X_subset, class_selected = ds.sample_n_from_same_class(n=n_series, split="train", seed=0)

print(f"Classe sélectionnée : {class_selected}")

# Initialisation du barycentre = moyenne euclidienne
barycenter_init = np.mean(X_subset, axis=0)

# FIGURE
plt.figure(figsize=(15, 4))

gammas = [0.01, 1.0]

for i, gamma in enumerate(gammas, start=1):

    # subplot
    ax = plt.subplot(1, 3, i)

    # tracer les séries originales
    for x in X_subset:
        ax.plot(x.ravel(), c="gray", linewidth=3, alpha=0.15)

    # calcul barycentre soft-DTW
    Z = softdtw_barycenter(X_subset, barycenter_init, gamma=gamma)

    # tracer barycentre
    ax.plot(Z.ravel(), c="red", linewidth=3, alpha=0.8)

    ax.set_title(f"Soft-DTW (γ={gamma})")

plt.tight_layout()
plt.show()
```
